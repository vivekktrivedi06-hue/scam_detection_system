```python
import seaborn as sns
titanic = sns.load_dataset("titanic")
```

 3.box plot 3rd topic
 

```python
sns.boxplot(x=titanic['sex'],y=titanic['age'],hue=titanic['survived'])
```

 4.distplot 4th topic

```python
sns.distplot(titanic[titanic['survived'] == 0]['age'], hist=False, label='Not Survived')

# Survived
sns.distplot(titanic[titanic['survived'] == 1]['age'], hist=False, label='Survived')

```

```python
# use kdeplot instead of distplot
#not survived
sns.kdeplot(titanic[titanic['survived'] == 0]['age'], label='Not Survived', fill=False)

# Survived
sns.kdeplot(titanic[titanic['survived'] == 1]['age'], label='Survived', fill=False)
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
# Sum only numeric columns
new=flights.groupby('year')[['passengers']].sum().reset_index()
```

```python
sns.lineplot(x=new['year'],y=new['passengers'])
```

```python
flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
sns.heatmap(flights.pivot_table(values='passengers',index='month',columns='year'))
```

```python
 sns.clustermap(flights.pivot_table(values='passengers',index='month',columns='year'))
```

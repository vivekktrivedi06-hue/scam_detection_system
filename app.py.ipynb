```python
import streamlit as st
import joblib

# Load model
model = joblib.load("phishing_pipeline.pkl")

st.set_page_config(
    page_title="Scam Detection System",
    layout="centered"
)

# -------- PROFESSIONAL SECURITY THEME -------- #
st.markdown("""
<style>
.stApp {
    background-color: #0B1F3A;
}

/* Title */
h1 {
    color: #FFFFFF;
    text-align: center;
    font-size: 40px;
    font-weight: 600;
}

/* Divider */
hr {
    border: 1px solid #1F4068;
}

/* Labels */
label {
    color: #DCE3F0 !important;
    font-weight: 500;
}

/* Text Area */
.stTextArea textarea {
    background-color: #132A4A;
    color: white;
    border-radius: 8px;
    border: 1px solid #1F4068;
}

/* Buttons */
.stButton>button {
    background-color: #1F4068;
    color: white;
    border-radius: 8px;
    height: 3em;
    width: 100%;
    font-weight: 600;
    border: none;
}

.stButton>button:hover {
    background-color: #2E5C9A;
}

/* Result Box */
.result-box {
    padding: 16px;
    border-radius: 8px;
    text-align: center;
    font-size: 18px;
    margin-top: 20px;
    font-weight: 500;
}
</style>
""", unsafe_allow_html=True)

# -------- TITLE -------- #
st.title("Scam Detection System")

st.markdown("---")

# -------- OPTION SELECT -------- #
option = st.radio(
    "Choose Content Type",
    ("Email", "Message")
)

st.markdown("---")

# -------- INPUT -------- #
user_input = st.text_area(
    f"Enter {option} Content Below:",
    height=200
)

# -------- ANALYZE BUTTON -------- #
if st.button("Analyze Content"):
    if user_input.strip() == "":
        st.warning("Please enter content to analyze.")
    else:
        prediction = model.predict([user_input])[0]

        if prediction == 1:
            st.markdown(
                '<div class="result-box" style="background-color:#3A0F17;color:#FF6B6B;">⚠ Scam Detected</div>',
                unsafe_allow_html=True
            )
        else:
            st.markdown(
                '<div class="result-box" style="background-color:#0F3A2A;color:#4BE3A3;">✔ Safe Content</div>',
                unsafe_allow_html=True
            )

st.markdown("---")
st.caption("Machine Learning Based Threat Classification")
```
